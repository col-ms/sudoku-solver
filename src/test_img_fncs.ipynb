```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
import read_puzzle
import cv2
import numpy as np
```

```python
folder = "../data/puzzle_imgs"

img_list = []
for file_name in os.listdir(folder):
    img_list.append(file_name)

puzzle_file = random.choice(img_list)
print(puzzle_file)
puzzle_img = cv2.imread(folder + "/" + puzzle_file)
plt.figure()
plt.imshow(puzzle_img)
plt.show()
```

```python
pre_proc_img = read_puzzle.pre_process_img(puzzle_img)
plt.figure()
plt.imshow(pre_proc_img)
plt.show()
```

```python
puzzle_bounds = read_puzzle.find_puzzle_bounds(pre_proc_img)
print(puzzle_bounds)
```

```python
new_puzzle_img = read_puzzle.crop_and_warp(pre_proc_img, puzzle_bounds)
plt.figure()
plt.imshow(new_puzzle_img)
plt.show()
```
